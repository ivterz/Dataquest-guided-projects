# Titanic: Machine Learning from Disaster

In this project, we will be designing a workflow in order to compete in the [Titanic: Machine Learning from Disaster](https://www.kaggle.com/c/titanic/overview) Kaggle competition.

[The sinking of the Titanic](https://en.wikipedia.org/wiki/Sinking_of_the_Titanic) one of the most infamous shipwrecks in history.

On April 15, 1912, during her maiden voyage, the widely considered “unsinkable” RMS Titanic sank after colliding with an iceberg. Unfortunately, there were not enough lifeboats for everyone onboard, resulting in the death of 1502 out of 2224 passengers and crew.

While there was some element of luck involved in surviving, it seems some groups of people were more likely to survive than others. In this project we will work with a dataset containing the passengers on the Titanic and we will try to create a Machine Learning model which will predict whether a passenger survived or not.

The data has been split into two groups:

- training set (`train.csv`)
- test set (`test.csv`)

The columns of the dataset are:

- `survival` - whether the passenger survived;	`0` = No, `1` = Yes
- `pclass` - ticket class;	`1` = 1st, `2` = 2nd, `3` = 3rd
- `sex` - the sex of the passenger
- `Age` - the age of the passenger
- `sibsp` - # of siblings/spouses aboard the Titanic
- `parch` - # of parents / children aboard the Titanic
- `ticket` - ticket number
- `fare` - passenger fare
- `cabin` - cabin number
- `embarked` - prt of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton

We can handle the missing values in `Fare` by populating them with the mean value of the column. Additionally, since there are so many unique values, we can create different bins which would include the price range in which the fare paid by the passenger falls.

Finally, since some of these columns will have to be converted to categorical data so that the machine can learn from them. Let's write functions which will help us make all the changes outlined above. We need to run them both on the `train` and `holdout` dataset.

```python
# %load functions.py
def process_missing(df):
    """Handle various missing values from the data set

    Usage
    ------

    holdout = process_missing(holdout)
    """
    df["Fare"] = df["Fare"].fillna(train["Fare"].mean())
    df["Embarked"] = df["Embarked"].fillna("S")
    return df

def process_age(df):
    """Process the Age column into pre-defined 'bins' 

    Usage
    ------

    train = process_age(train)
    """
    df["Age"] = df["Age"].fillna(-0.5)
    cut_points = [-1,0,5,12,18,35,60,100]
    label_names = ["Missing","Infant","Child","Teenager","Young Adult","Adult","Senior"]
    df["Age_categories"] = pd.cut(df["Age"],cut_points,labels=label_names)
    return df

def process_fare(df):
    """Process the Fare column into pre-defined 'bins' 

    Usage
    ------

    train = process_fare(train)
    """
    cut_points = [-1,12,50,100,1000]
    label_names = ["0-12","12-50","50-100","100+"]
    df["Fare_categories"] = pd.cut(df["Fare"],cut_points,labels=label_names)
    return df

def process_cabin(df):
    """Process the Cabin column into pre-defined 'bins' 

    Usage
    ------

    train process_cabin(train)
    """
    df["Cabin_type"] = df["Cabin"].str[0]
    df["Cabin_type"] = df["Cabin_type"].fillna("Unknown")
    df = df.drop('Cabin',axis=1)
    return df

def process_titles(df):
    """Extract and categorize the title from the name column 

    Usage
    ------

    train = process_titles(train)
    """
    titles = {
        "Mr" :         "Mr",
        "Mme":         "Mrs",
        "Ms":          "Mrs",
        "Mrs" :        "Mrs",
        "Master" :     "Master",
        "Mlle":        "Miss",
        "Miss" :       "Miss",
        "Capt":        "Officer",
        "Col":         "Officer",
        "Major":       "Officer",
        "Dr":          "Officer",
        "Rev":         "Officer",
        "Jonkheer":    "Royalty",
        "Don":         "Royalty",
        "Sir" :        "Royalty",
        "Countess":    "Royalty",
        "Dona":        "Royalty",
        "Lady" :       "Royalty"
    }
    extracted_titles = df["Name"].str.extract(' ([A-Za-z]+)\.',expand=False)
    df["Title"] = extracted_titles.map(titles)
    return df

def create_dummies(df,column_name):
    """Create Dummy Columns (One Hot Encoding) from a single Column

    Usage
    ------

    train = create_dummies(train,"Age")
    """
    dummies = pd.get_dummies(df[column_name],prefix=column_name)
    df = pd.concat([df,dummies],axis=1)
    return df
```

```python
def process_all(df):
    
    """
    Runs all the functions from above.
    """
    
    for f in [process_missing, process_age, process_fare, process_cabin, process_titles]:
        df = f(df)
        
    for c in ["Age_categories", "Fare_categories","Title", "Cabin_type", "Sex"]:
        df = create_dummies(df, c)
        df = df.drop(c, axis=1)
        
    return df
```

```python
train = process_all(train)
holdout = process_all(holdout)
```

```python
train = train.drop(['Age', 'Fare'], axis=1)
holdout = holdout.drop(['Age', 'Fare'], axis=1)
```

We see that about 30% of the passengers who did not have any family members on board survived. Of those who had 7+ family members onboard, none survived.

Let's create a new feature in the dataframe which will indicate whether the passenger was alone onboard:

- `1` - if the passenger had zero family members onboard
- `0` - if the passenger had one or more family members onboard

```python
def accompanied(df):
    """
    Creates a `isalone` column based on the number of family members onboard
    """
    
    df['isalone'] = df['family_members'].apply(lambda x: 0 if x == 0 else 1)
    df = df.drop(['family_members','SibSp', 'Parch'], axis=1)
    return df
```

```python
train = accompanied(train)
holdout = accompanied(holdout)
```

Now, let's write a function which will help us identify the best columns to use for our predictions.

```python
def select_features(df):
    df = df.select_dtypes(include='number')
    df = df.dropna(axis=1)
            
    all_X = df.drop(['PassengerId', 'Survived'], axis=1)
    all_y = df['Survived']
    
    rfc = RandomForestClassifier(random_state=1)
    selector = RFECV(rfc, cv=10)
    selector.fit(all_X, all_y)
    
    best_columns = all_X.columns[selector.support_]
    return list(best_columns)
    
```

```python
best_columns = select_features(train)
best_columns
```

Now, let's find out which combination machine learning model and hyprparameters would be best for our predictions.

```python
def select_model(df, features):
    all_X = df[features]
    all_y = df['Survived']
    
    models = [
        {
        'name': 'LogisticRegression',
        'estimator': LogisticRegression(),
        'hyperparameters': 
            {
            'solver': ["newton-cg", "lbfgs", "liblinear"]
            }
        },
        {
        'name': 'KNeighborsClassifier',
        'estimator': KNeighborsClassifier(),
        'hyperparameters': 
            {
            "n_neighbors": range(1,20,2),
            "weights": ["distance", "uniform"],
            "algorithm": ["ball_tree", "kd_tree", "brute"],
            "p": [1,2]
            } 
        },
        {
        'name': 'RandomForestClassifier',
        'estimator': RandomForestClassifier(),
        'hyperparameters': 
            {
            "n_estimators": [4, 6, 9],
            "criterion": ["entropy", "gini"],
            "max_depth": [2, 5, 10],
            "max_features": ["log2", "sqrt"],
            "min_samples_leaf": [1, 5, 8],
            "min_samples_split": [2, 3, 5]          
            }
        }
    ]  
    
    for m in models:
        print(m['name'])
        print(12*'-')
        grid = GridSearchCV(m['estimator'], param_grid=m['hyperparameters'], cv=10)
        grid.fit(all_X, all_y)
        m['best_params'] = grid.best_params_
        m['best_score'] = grid.best_score_
        m['best_model'] = grid.best_estimator_
        print('Best score:', grid.best_score_)
        print('Best parameters:', grid.best_params_)
        print('\n')
    return models
```

```python
models = select_model(train, best_columns)
```

```python
def save_submission_file(model, cols, filename='submission.csv'):
    predictions = model.predict(holdout[cols])
    submission = pd.DataFrame(holdout['PassengerId'])
    submission['Survived'] = predictions
    submission.to_csv(filename, index=False)

best_model = models[2]['best_model']
save_submission_file(best_model, best_columns, 'submission_2.csv')
```
